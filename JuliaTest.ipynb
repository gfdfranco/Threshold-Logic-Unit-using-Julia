```julia
#Libraries
using CSV
using DataFrames
using Plots
plotly()
```

## Hyperparameters

```julia
weights = [0.5,0.5,0.5]
features = 3 
e = 0.1
t = 0.1
epochs = 1000
```

```julia
function reset_values()
    weights = [0.5,0.5,0.5]
    features = 3 
    e = 0.1
    t=0.1
    epochs=1000
    return nothing
end
```

## Delta rule algorithm 

```julia
function train(df,dependent_variable_column_name)
    global t
    for i in 1:epochs
        number_row = rand((1:8))
        row = df[number_row,["x1", "x2", "x3",dependent_variable_column_name]]
        X = (row["x1"], row["x2"], row["x3"])
        y = row[dependent_variable_column_name]
        E_temp = weights[1]*X[1] + weights[2]*X[2] + weights[3]*X[3] 

        if y == 1
            if E_temp < t
                E = abs(abs(t) - abs(E_temp))
                d = (E + e)/2
                t = t - d
                for i in 1:features
                    if X[i] == 1
                        weights[i] = weights[i] + d   
                    end
                end
            end
        else
            if E_temp > t
                E = abs(abs(t) - abs(E_temp))
                d = (E + e)/2
                t = t + d
                for i in 1:features
                    if X[i] == 1
                        weights[i] = weights[i] - d   
                    end
                end
            end
        end
    end
    return nothing
end
```

## Plot results

```julia
function plot_results(df, dependent_variable_column_name)
    x = df[:, "x1"]
    y = df[:, "x2"]
    z = df[:, "x3"]
    x2=range(0,stop=1,length=2)
    y2=range(0,stop=1,length=2)
    f(x2,y2) = (-1*(weights[1]*x2+weights[2]*y2))/weights[3]
    dependent_variable =  df[:, dependent_variable_column_name]
    p = Plots.scatter3d(x, y, z,   st = [:scatter3d], color = dependent_variable) 
    return plot!(p, x2, y2, f, st=:surface)
end
```

```julia
reset_values()
train(data_ex1, "f2")

#Get weights for F2
for i in 1:features
    println(weights[i])
end
```
